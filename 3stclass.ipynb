### **string、list、tuple and set 的内置函数和方法全总结**
---
**String（不可变，常用方法）**
- 常用方法：lower(), upper(), strip(), split(sep), join(iterable), replace(old,new), find(), index(), startswith(), endswith(), isdigit()
- 典型场景：文本清洗、切分、格式化输出
```python
s = '  Hello,World  '
print(s.strip().lower())  # 'hello,world'
```

**List（可变、有序）**
- 常用方法：append(), extend(), insert(), pop([i]), remove(x), clear(), index(x), count(x), sort(key=...), reverse(), copy()（浅拷贝）
- 典型场景：栈/队列、就地修改、结果累积
```python
a = [3,1,2]
a.sort()   # 就地排序 => [1,2,3]
b = a.copy()  # 避免副作用
```

**Tuple（不可变、有序）**
- 常用方法：count(), index()；用途：拆包、函数返回、可作为 dict 的键（元素都可哈希时）
- 典型场景：不希望被修改的记录/配置
```python
t = (1,2,3)
x,y,z = t  # 拆包
```

**Set（可变、无序、去重）**
- 常用方法：add(), remove()/discard(), pop(), clear(), update(), union(|), intersection(&), difference(-), symmetric_difference(^), issubset()/issuperset()
- 典型场景：去重、集合运算、成员测试（平均 O(1)）
```python
s = {1,2,3}
s.add(4)
print(2 in s)  # True
```

````python
# 示例：string / list / tuple / set 的常见用法和输出
s = '  Hello,World  '
print('strip lower ->', s.strip().lower())

a = [3,1,2]
a.sort()
print('list sort ->', a)
b = a.copy()
b.append(4)
print('original list after copy and append to copy ->', a, b)

t = (1,2,3)
x,y,z = t
print('tuple unpack ->', x,y,z)

st = {1,2,3}
st.add(4)
print('set contains 2? ->', 2 in st)
````

````python
# 示例：通用内置函数演示
arr = [5,2,9]
print('len, min, max, sum ->', len(arr), min(arr), max(arr), sum(arr))
print('sorted (new) ->', sorted(arr))
print('enumerate ->', list(enumerate(arr)))
print('any/all on [True, False] ->', any([True, False]), all([True, False]))

# map / filter 示例
print('map x*2 ->', list(map(lambda x: x*2, arr)))
print('filter even ->', list(filter(lambda x: x%2==0, arr)))
````

### **文件读写 file operator**
---
#### 打开文件：模式与编码
- open(path, mode='r', encoding=None)：常用 mode：
  - 'r' 只读（默认）；'w' 写（会覆盖）；'a' 追加；'x' 独占创建；
  - 在模式后加 'b' 表示二进制（如 'rb','wb'）；在文本模式中可指定 encoding='utf-8'；

#### 推荐：上下文管理器（with）
- 使用 with 可以自动关闭文件并在异常时安全处理：
```python
with open('data.txt', 'r', encoding='utf-8') as f:
    text = f.read()
# 无需手动 f.close()
```

#### 读取方法对比
- f.read(size=-1)：读取整个文件或指定字节数/字符数；
- f.readline()：读取一行；f.readlines()：返回所有行的列表（慎用大文件）；
- 逐行遍历（最常用、内存友好）：
```python
with open('big.txt','r',encoding='utf-8') as f:
    for line in f:
        process(line)
```

#### 写入方法
- f.write(text)：写入字符串（在文本模式）；f.writelines(list_of_strings) 写入多行（不自动加换行）；
- 使用 flush() 强制刷新缓冲区，通常不需要（with 会自动 flush/close）；

#### 文件指针操作：seek / tell
- f.seek(offset, whence)：移动文件指针；f.tell() 返回当前偏移；常用于二进制处理或随机访问。

#### 二进制文件处理
- 使用 'rb' / 'wb' 读取/写入字节，如图片、音频、序列化文件。
```python
with open('img.png','rb') as fr, open('copy.png','wb') as fw:
    data = fr.read()
    fw.write(data)
```
