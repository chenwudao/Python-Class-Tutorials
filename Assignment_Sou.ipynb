```python
def transform_string(s):
    # Step 1: 交替转换非空格字符大小写
    step1 = []
    non_space_count = 0  # 1-based计数非空格字符
    for char in s:
        if char != ' ':
            non_space_count += 1
            # 奇数位大写，偶数位小写
            step1.append(char.upper() if non_space_count % 2 == 1 else char.lower())
        else:
            step1.append(char)  # 空格保留原位
    step1_str = ''.join(step1)
    
    # Step 2: 反转单词顺序，保留空格结构（不用 re）
    tokens = []  # 存储单词或空格段
    current_token = []
    is_space = None  # 标记当前在处理空格还是单词

    for char in step1_str:
        char_is_space = (char == ' ')
        
        # 状态切换时保存当前 token
        if is_space is not None and is_space != char_is_space:
            tokens.append(''.join(current_token))
            current_token = []
        
        current_token.append(char)
        is_space = char_is_space

    # 处理最后一个 token
    if current_token:
        tokens.append(''.join(current_token))

    # 提取所有单词并反转
    words = [token for token in tokens if token and token[0] != ' ']
    reversed_words = words[::-1]

    # 重构字符串
    step2 = []
    word_idx = 0
    for token in tokens:
        if token and token[0] == ' ':  # 空格段
            step2.append(token)
        else:  # 单词段
            step2.append(reversed_words[word_idx])
            word_idx += 1

    return ''.join(step2)

if __name__ == "__main__":
    input_str = input()
    print(transform_string(input_str))
```

```python
n = int(input().strip())
table = []
for _ in range(n + 1):
    line = input().strip('\n')  # 保留原始逗号分隔结构
    table.append(line.split(','))

# 读取操作数并处理
m = int(input().strip())
for _ in range(m):
    line = input().strip('\n')
    if not line:
        continue
    
    # 解析操作类型
    if line.startswith('append '):
        op = 'append'
        values = line[len('append '):]  # 提取append后的所有内容
    else:
        parts = line.split()
        if len(parts) == 0:
            continue
        op = parts[0]
    
    # 处理filter操作
    if op == 'filter':
        if len(parts) < 2 or '=' not in parts[1]:
            continue
        col_name, val = parts[1].split('=', 1)
        if col_name not in table[0]:
            continue  # 无效列名，忽略操作
        col_idx = table[0].index(col_name)
        # 过滤数据行（保留匹配项）
        filtered_data = [row for row in table[1:] if row[col_idx] == val]
        table = [table[0]] + filtered_data
    
    # 处理append操作
    elif op == 'append':
        new_row = values.split(',')
        if len(new_row) == len(table[0]):  # 列数匹配才添加
            table.append(new_row)
    
    # 处理delete操作
    elif op == 'delete':
        if len(parts) < 2:
            continue
        try:
            index = int(parts[1])  # 1-based索引
            data_idx = index - 1  # 转换为0-based数据行索引
        except ValueError:
            continue  # 非整数索引，忽略
        # 检查索引有效性（数据行范围：0 ~ 数据行数-1）
        if 0 <= data_idx < len(table[1:]):
            del table[data_idx + 1]  # 表头占索引0，数据行从1开始

# 生成CSV格式输出
csv_output = '\n'.join([','.join(row) for row in table])
print(csv_output)
```
